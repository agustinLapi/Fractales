```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def hurst(x_i,x_f,D):
  H = 2 - D
  x_m = ( x_i + x_f )/2
  y_m = np.random.normal(0, x_m ** H)
  return y_m

#n_points must be a power of 2
def fbm(D, n_points = 1024):
  n_points = 1024
  t = np.linspace(0,1,n_points,endpoint=True) #time vector
  v = np.zeros((1,n_points))
  vector_len = int(np.log2(n_points))
  v[0,0] = random.random()
  v[0,-1] = random.random()
  # for y vector I need to use hurst function
  step = np.ones((1,vector_len))
  for i in range(vector_len):
    step[0,i] = 1/(2**(i+1))
  #print(step)
  for i in step[0,:] :
    aux = int(n_points*i)
    #print(aux)
    for j in range(0, n_points, aux):
      #print(j)
      if (v[0,j]) == 0:
        v[0,j] = hurst(t[j-aux],t[j+aux-1],D)
  return t, v
```

```python
x,y = fbm(0.3)
plt.figure
plt.plot(x[:],y[0,:])
```
