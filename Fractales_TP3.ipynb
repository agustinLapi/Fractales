```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ec_log(x, alfa, iteraciones ):
  x_i = []
  x_i.append(x)
  for i in range(iteraciones):
    aux_x = x_i[-1]
    x_i.append(alfa*aux_x*(1-aux_x))

  plt.figure()
  jj = np.arange(len(x_i))
  plt.plot(jj, x_i, 'r')

  plt.show()
  if alfa <=1:
    print("Population dies")
  elif alfa > 1 and alfa <=3:
    convergence_point=(alfa-1)/alfa
    print("Converges to ", convergence_point)
  elif alfa >3 and alfa <3.44949 :
    print("Oscillates between ", min(x_i[-100:])," and ", max(x_i[-100:]) )
  return
```

```python
def ec_log_2(x, alfa, iteraciones):
  x_i = []
  x_i.append(x)
  for i in range(iteraciones):
    aux_x = x_i[-1]
    x_i.append(alfa*aux_x*(1-aux_x))

  return x_i[-2], x_i[-1]

def poincare(alfa, step, iteraciones, seeds, xlim):
  plt.figure()
  _i = int((alfa[1]-alfa[0])//step)
  print(alfa[0])
  for ii in range(_i):
    for jj in range(seeds):
      x = np.random.normal(0.5, 0.2)
      while(x>1 and x<0):
        x = np.random.normal(0.5, 0.2)  #x must be between 0 and 1
      for kk in range(iteraciones):
        my_alfa= alfa[0]+ii*step
        aux_1, aux_2 = ec_log_2(x, my_alfa, iteraciones)
      #print(aux_1)
      plt.plot(my_alfa, aux_1, marker=".", markersize=1, markerfacecolor="black")
      plt.plot(my_alfa, aux_2, marker=".", markersize=1, markerfacecolor="black")

  plt.xlim(xlim[0],xlim[1])
  plt.show()
  return
```
