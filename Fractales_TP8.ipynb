```python
import numpy as np
import matplotlib.pyplot as plt
from ColabTurtle.Turtle import *
import random
```

```python
def Barnsley(iter, A, B, C, D, E, F, P):
  x = 0
  y = 0
  x_coor = []
  y_coor = []
  for n in range(iter):
    #goto(65 * x + 350, -37 * y + 410)
    #x = 65 * x + 350
    #y = -37 * y + 410
    r = random.random()
    xn = x
    yn = y
    if r <= P[0]:
        x = A[0] * xn + B[0] * yn + E[0]
        y = C[0] * xn + D[0] * yn + F[0]
    elif r <= (P[0]+P[1]):
        x = A[1] * xn + B[1] * yn + E[1]
        y = C[1] * xn + D[1] * yn + F[1]
    elif r <= (P[0]+P[1]+P[2]):
        x = A[2] * xn + B[2] * yn + E[2]
        y = C[2] * xn + D[2] * yn + F[2]
    else:
        x = A[3] * xn + B[3] * yn + E[3]
        y = C[3] * xn + D[3] * yn + F[3]
    x_coor.append(x)
    y_coor.append(y)        
  return x_coor, y_coor
```

```python
initializeTurtle()
speed(12)
width(2)
penup()
hideturtle()
x = 0
y = 0
for n in range(110000):
    goto(65 * x + 350, -37 * y + 410)
    pendown()
    forward(0.2)
    backward(0.2)
    penup()
    r = random.random()
    r = r * 100
    xn = x
    yn = y
    if r < 1:
        x = 0
        y = 0.16 * yn
    elif r < 86:
        x = 0.85 * xn + 0.04 * yn
        y = -0.04 * xn + 0.85 * yn + 1.6
    elif r < 93:
        x = 0.20 * xn - 0.26 * yn
        y = 0.23 * xn + 0.22 * yn + 1.6
    else:
        x = -0.15 * xn + 0.28 * yn
        y = 0.26 * xn + 0.24 * yn + 0.44
```

```python
itr = 100000
A = [0,  0.85,  0.20, -0.15]
B = [0,  0.04, -0.26,  0.28]
C = [0, -0.04,  0.23,  0.26]
D = [0.16, 0.85, 0.22, 0.24]
E = [0, 0, 0, 0]
F = [0, 1.6, 1.6, 0.44]
P = [0.01, 0.85, 0.07, 0.07]
X,Y = Barnsley(itr, A, B, C, D, E, F, P)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X,Y, s=0.4)
```
