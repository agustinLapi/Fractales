```python
import cmath as cm 
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def inversa(zo,C, iter):
  res = []
  res.append(zo)
  for i in range(iter):
    zi = cm.sqrt(res[-1]-C)
    if random.random() < 0.5:
      res.append( zi )
    else:
      res.append( -zi )
  # extract real part
  x = [ele.real for ele in res]
  # extract imaginary part
  y = [ele.imag for ele in res]
  return x,y


def sembrar(alfa = 0.5):
  return np.random.normal(0, alfa) + 1j*np.random.normal(0, alfa)

def mandelbrot(iter):
  res = []
  zo = 0
  C = sembrar()
  res.append(zo)
  for i in range(iter):
    zi = res[-1]*res[-1]+C
    res.append( zi )
  # extract real part
  x = [ele.real for ele in res]
  # extract imaginary part
  y = [ele.imag for ele in res]

  return x, y

def semillas(iter, C):
  res = []
  zo = sembrar()
  res.append(zo)
  for i in range(iter):
    zi = res[-1]*res[-1]+C
    res.append( zi )
  # extract real part
  x = [ele.real for ele in res]
  # extract imaginary part
  y = [ele.imag for ele in res]
  return x,y

def preimagen(zo, C, iter):
  res = []
  res.append(zo)
  for i in range(iter):
    zi = res[-1]*res[-1]+C
    res.append( zi )
  # extract real part
  x = [ele.real for ele in res]
  # extract imaginary part
  y = [ele.imag for ele in res]
  return x,y
```

```python
x, y = inversa(1,0,10000)  
# plot the complex numbers
plt.scatter(x, y,s=1)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.show()
```
